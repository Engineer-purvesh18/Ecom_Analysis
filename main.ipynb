## DATA CLEANING

```python
# Dropping the postal code column as it has very high number of missing values
df_orders = df_orders.drop(columns=['Postal Code'])
```

```python
# Formatting the Order Date and Ship Date columns in proper date time format
df_orders['Order Date'] = pd.to_datetime(df_orders['Order Date'])
df_orders['Ship Date'] = pd.to_datetime(df_orders['Ship Date'])
```

```python
# converting yes and no in returned column in df_return to binary input 
df_returns['Returned'] = df_returns['Returned'].apply(lambda x: 1 if x == 'Yes' else 0) # Yes = 1, NO = 0
```
