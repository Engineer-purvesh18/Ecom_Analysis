```python
# Importing required libraries and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Import sheets 
sheets = pd.read_excel("ECOMM DATA.xlsx", sheet_name=None)
```

```python
# Assigning names and creating diffferent data frames form sheets in the xlsl file
df_orders = sheets['Orders']
df_returns = sheets['Returns']
df_people = sheets['People']
```

```python
# Getting Basic overview of the tables.
df_orders.head(5)
```

```python
# Getting number of null values form all three data frames
for name, df in {
    'Orders': df_orders,
    'Returns': df_returns,
    'People': df_people
}.items():
    print(f"{name}:\n{df.isnull().sum()}\n")
```
